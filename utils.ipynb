```python
import pandas as pd
from elasticsearch import Elasticsearch, helpers

# Initialiser le client Elasticsearch
es = Elasticsearch(["https://localhost:9200"], basic_auth=('elastic', 'changeme'), verify_certs=False)

# Lire le fichier CSV
file_path = 'bano-25.csv'  # Changez ce chemin si ncessaire
data = pd.read_csv(file_path, header=None)

# Renommer les colonnes pour faciliter la manipulation
data.columns = ['id', 'num', 'street', 'postal_code', 'city', 'source', 'latitude', 'longitude']

# Fonction pour transformer les donnes au format Elasticsearch
def generate_actions(data):
    for _, row in data.iterrows():
        yield {
            "_index": "adresses_geo",  # Utiliser le nouvel index avec geo_point
            "_id": row['id'],  # Identifiant unique pour chaque document
            "_source": {
                "num": row['num'],
                "street": row['street'],
                "postal_code": row['postal_code'],
                "city": row['city'],
                "source": row['source'],
                "location": {  # Utiliser le format geo_point
                    "lat": row['latitude'],
                    "lon": row['longitude']
                }
            }
        }

# Indexer les donnes dans Elasticsearch
try:
    helpers.bulk(es, generate_actions(data))
    print("Les donnes ont t indexes avec succs.")
except Exception as e:
    print(f"Une erreur s'est produite lors de l'indexation : {e}")
```
